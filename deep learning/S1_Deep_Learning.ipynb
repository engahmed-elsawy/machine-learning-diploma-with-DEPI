```python
import numpy as np
# activation funcation (sigmoid)
def activation(z):
    return 1 / (1 + np.exp(-z))

# Derivative activation funcation for backprop
def activation_derivative(z):
    return activation(z) * (1 - activation(z))
```

```python
# Perceptron and apply activation funcation
def perceptron(inputs, weights, bias):
    weighted_sum = np.dot(inputs, weights) + bias
    return activation(weighted_sum)
```

```python
# compute loss (mse ==> Mean Square Error)
def compute_loss(y_pred, y_true):
    return np.mean((y_pred - y_true) ** 2)
```

```python
# update weights and bias
def update_weights(inputs, weights, bias, y_pred, y_true, learning_rate = 0.01):
    error = compute_loss(y_pred, y_true) # calc the error
    weights_gradient = np.dot(inputs.T, error * activation_derivative(y_pred)) # weights_gradient
    bias_gradient = np.sum(error * activation_derivative(y_pred)) # bias_gradient

    weights -= learning_rate * weights_gradient  # update weights
    bias -= learning_rate * bias_gradient  # update bias

    return weights, bias
```

```python
inputs = np.random.randint(0, 10, (3,2))  # 3,2
weights = np.random.rand(2)
bias = np.random.rand(1)
y_true = np.array([0,1,0])
```

```python
y_pred = perceptron(inputs, weights, bias) # forward
loss = compute_loss(y_pred, y_true) # copmute loss

# epchos for backprop
for epcho in range(50):
    weights, bias = update_weights(inputs, weights, bias, y_pred, y_true, learning_rate = 0.01)
    y_pred = perceptron(inputs, weights, bias) # forward
    loss = compute_loss(y_pred, y_true)
    print(f"loss for epcho {epcho} and the loss is {loss}")
```
