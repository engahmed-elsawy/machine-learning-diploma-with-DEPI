```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
```

## task 2

```python
df = sns.load_dataset("titanic")
```

## task 3

```python
df.head()
```

## task 4

```python
df.head()
```

## task 5

```python
df.describe()
```

```python
df.describe(include="object")
```

## task 6

```python
df.isnull().sum()
```

## task 7

```python
df["age"].fillna(df["age"].median(), inplace=True)
```

## task 8

```python
df["embarked"].fillna(df["embarked"].mode()[0], inplace=True)
```

 ## task 9

```python
df.drop(columns=["deck"], inplace=True)
```

 ## task 10

```python
df.dropna(subset=["embark_town"], inplace=True)
```

## task 11

```python
df.isnull().sum()
```

## task 12

```python
plt.hist(df["age"], bins=30)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

## task 13

```python
sns.countplot(data=df, x="sex", hue="survived")
plt.title("Survival Count by Gender")
plt.show()
```

## task 14

```python
num_cols = df.select_dtypes(include=["int64", "float64"])
corr = num_cols.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## task 15

```python
sns.barplot(data=df, x="pclass", y="survived")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

## task 16

```python
sns.boxplot(data=df, x="survived", y="fare")
plt.title("Fare Distribution by Survival Status")
plt.show()
```

## task 17

```python
df["family_size"] = df["sibsp"] + df["parch"]
df["is_alone"] = np.where(df["family_size"] == 0, 1, 0)
```

## task 18

```python
df[["sibsp", "parch", "family_size", "is_alone"]].head()
```

## task 19


```python
sns.barplot(data=df, x="family_size", y="survived")
plt.title("Survival Rate by Family Size")
plt.show()
```

## task 20

```python
sns.countplot(data=df, x="embark_town")
plt.title("Passengers by Embark Town")
plt.xticks(rotation=45)
plt.show()
```

## task 21


```python
sns.barplot(data=df, x="embark_town", y="survived")
plt.title("Survival Rate by Embark Town")
plt.xticks(rotation=45)
plt.show()
```

## task 22

```python
sns.boxplot(data=df, x="pclass", y="age")
plt.title("Age Distribution by Passenger Class")
plt.show()
```

```python
sns.boxplot(data=df, x="pclass", y="fare")
plt.title("Fare Distribution by Passenger Class")
plt.show()
```
