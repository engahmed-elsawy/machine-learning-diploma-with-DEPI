# Python_Session 1: Data Structures & Control Flow

```python
fruits = ["Apple", "banana", "Orange", "apple", "grape", "Kiwi", "Banana", "mango"]
print(f"Original list: {fruits}")

#Add 2 fruits only if they aren't already in the list (case-insensitive)
new_fruits_to_add = ["orange", "Blueberry"]
for new_fruit in new_fruits_to_add:
    current_lower = [f.lower() for f in fruits]

    if new_fruit.lower() not in current_lower:
        fruits.append(new_fruit)
        print(f"Added: {new_fruit}")
    else:
        print(f"Skipped (duplicate): {new_fruit}")
#Remove all occurrences of a fruit (case-insensitive)
target_to_remove = "banana"
fruits = [f for f in fruits if f.lower() != target_to_remove.lower()]
print(f"After removing '{target_to_remove}': {fruits}")
#New list of fruits starting with a vowel (List Comprehension)
vowels = "aeiou"
vowel_fruits = [f for f in fruits if f[0].lower() in vowels]
print(f"Starts with vowel: {vowel_fruits}")
#Sort alphabetically ignoring case
fruits.sort(key=str.lower)
print(f"Sorted list: {fruits}")
```

```python
numbers = (45, 12, 88, 3, 99, 21)
print(f"\nOriginal Tuple: {numbers}")

#Print index of largest value (Manual Loop)
max_val = numbers[0]
max_index = 0

for i, num in enumerate(numbers):
    if num > max_val:
        max_val = num
        max_index = i

print(f"Largest value is {max_val} at index: {max_index}")

#Replace a value (Immutability Workaround)
replace_index = 2
new_value = 100

#Slice before + new value + slice after
new_tuple = numbers[:replace_index] + (new_value,) + numbers[replace_index+1:]

print(f"New Tuple: {new_tuple}")
```

```python
list_a = [1, 2, 2, 3, 4]
list_b = [3, 4, 5, 5, 6]

set_a = set(list_a)
set_b = set(list_b)

print(f"\nSet A: {set_a}")
print(f"Set B: {set_b}")

#Set Operations
print(f"Union (A | B): {set_a | set_b}")
print(f"Intersection (A & B): {set_a & set_b}")
print(f"Difference (A - B): {set_a - set_b}") # Items in A that are not in B

#Subset Check
small_set = {1, 2}
print(f"Is {small_set} a subset of A? {small_set.issubset(set_a)}")

#Remove duplicates preserving order
def clean_list_ordered(input_list):
    seen = set()
    result = []
    for item in input_list:
        if item not in seen:
            result.append(item)
            seen.add(item)
    return result

messy_list = [10, 20, 10, 30, 20, 40, 10]
cleaned = clean_list_ordered(messy_list)
print(f"Ordered Deduplication: {messy_list} -> {cleaned}")
```

```python
print("\n--- Word Frequency Counter ---")
text = "Python is great. Python is dynamic, and python is fun!"

clean_text = text.lower()
punctuation = ",.!?;:"
for char in punctuation:
    clean_text = clean_text.replace(char, "")

words = clean_text.split()

counts = {}
for w in words:
    counts[w] = counts.get(w, 0) + 1
most_common_word = ""
highest_count = 0

for word, count in counts.items():
    if count > highest_count:
        highest_count = count
        most_common_word = word

print(f"Word Counts: {counts}")
print(f"Most frequent: '{most_common_word}' (appears {highest_count} times)")
```

```python
def add(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError(f"Inputs must be numbers. Received: {type(a).__name__}, {type(b).__name__}")

    return a + b

try:
    print(add(10, 5.5))
    print(add(10, "5"))
except TypeError as e:
    print(f"Error caught: {e}")
```

```python
def divide_int(a: int, b: int):
    if b == 0:
        raise ZeroDivisionError("Math Error: Cannot divide by zero.")

    quotient = a // b
    remainder = a % b
    return quotient, remainder

try:
    q, r = divide_int(10, 3)
    print(f"10 divided by 3 is {q} with a remainder of {r}")

    divide_int(5, 0)
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
def factorial(n):
    if not isinstance(n, int) or n < 0:
        raise ValueError("Factorial is defined for non-negative integers only.")

    if n <= 1:
        return 1

    return n * factorial(n - 1)

def fibonacci(n, memo={}):
    if n in memo:
        return memo[n]

    if n <= 1:
        return n


    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)
    return memo[n]

print(f"Factorial of 5: {factorial(5)}")
print(f"Fibonacci of 50 (fast): {fibonacci(50)}")
```

# Python_Session 4: Object-Oriented Programming

```python
import json
import datetime
import uuid
import os

# ==========================================
# 1. Domain Classes
# ==========================================

class Room:
    VALID_TYPES = {"single", "double", "suite"}

    def __init__(self, room_number: int, room_type: str, price_per_night: float, max_guests: int):
        if room_type not in self.VALID_TYPES:
            raise ValueError(f"Invalid room type. Must be one of {self.VALID_TYPES}")
        if price_per_night <= 0:
            raise ValueError("Price must be > 0")
        if max_guests < 1:
            raise ValueError("Max guests must be >= 1")

        self.room_number = room_number
        self.room_type = room_type
        self.price_per_night = price_per_night
        self.max_guests = max_guests

    def to_dict(self):
        return {
            "room_number": self.room_number,
            "room_type": self.room_type,
            "price_per_night": self.price_per_night,
            "max_guests": self.max_guests
        }

    @classmethod
    def from_dict(cls, data):
        return cls(
            room_number=data["room_number"],
            room_type=data["room_type"],
            price_per_night=data["price_per_night"],
            max_guests=data["max_guests"]
        )

    def __str__(self):
        return f"[Room {self.room_number}] {self.room_type.capitalize()} - ${self.price_per_night}/night (Max: {self.max_guests})"


class Guest:
    def __init__(self, guest_id: str, name: str, phone: str):
        if not name or not phone:
            raise ValueError("Name and phone cannot be empty")
        self.guest_id = guest_id
        self.name = name
        self.phone = phone

    def to_dict(self):
        return {
            "guest_id": self.guest_id,
            "name": self.name,
            "phone": self.phone
        }

    @classmethod
    def from_dict(cls, data):
        return cls(data["guest_id"], data["name"], data["phone"])

    def __str__(self):
        return f"[Guest] {self.name} (ID: {self.guest_id}, Phone: {self.phone})"


class Reservation:
    VALID_STATUSES = {"booked", "checked_in", "checked_out", "cancelled"}

    def __init__(self, reservation_id: str, guest_id: str, room_number: int,
                 check_in: datetime.date, check_out: datetime.date, status: str = "booked"):
        if check_out <= check_in:
            raise ValueError("Check-out must be after check-in")
        if status not in self.VALID_STATUSES:
            raise ValueError(f"Invalid status: {status}")

        self.reservation_id = reservation_id
        self.guest_id = guest_id
        self.room_number = room_number
        self.check_in = check_in
        self.check_out = check_out
        self.status = status

    def nights(self):
        delta = self.check_out - self.check_in
        return delta.days

    def total_cost(self, price_per_night: float):
        return self.nights() * price_per_night

    def overlaps(self, other_check_in: datetime.date, other_check_out: datetime.date):
        """
        Logic: A overlaps B if (StartA < EndB) and (StartB < EndA)
        We treat check-out day as free for the next guest (half-open interval).
        """
        if self.status in ("cancelled", "checked_out"):
            return False

        return (self.check_in < other_check_out) and (other_check_in < self.check_out)

    def to_dict(self):
        return {
            "reservation_id": self.reservation_id,
            "guest_id": self.guest_id,
            "room_number": self.room_number,
            "check_in": self.check_in.isoformat(),
            "check_out": self.check_out.isoformat(),
            "status": self.status
        }

    @classmethod
    def from_dict(cls, data):
        c_in = datetime.date.fromisoformat(data["check_in"])
        c_out = datetime.date.fromisoformat(data["check_out"])
        return cls(
            reservation_id=data["reservation_id"],
            guest_id=data["guest_id"],
            room_number=data["room_number"],
            check_in=c_in,
            check_out=c_out,
            status=data["status"]
        )

    def __str__(self):
        return (f"[{self.status.upper()}] Res ID: {self.reservation_id} | Room: {self.room_number} | "
                f"Guest: {self.guest_id} | Dates: {self.check_in} to {self.check_out}")


# ==========================================
# 2. Hotel Controller
# ==========================================

class Hotel:
    def __init__(self):
        self.rooms = {}
        self.guests = {}
        self.reservations = {}

    def add_room(self, room: Room):
        if room.room_number in self.rooms:
            raise ValueError(f"Room {room.room_number} already exists.")
        self.rooms[room.room_number] = room

    def register_guest(self, guest: Guest):
        self.guests[guest.guest_id] = guest

    def search_available_rooms(self, check_in, check_out, room_type=None, guests_count=None):
        available = []
        for room in self.rooms.values():
            if room_type and room.room_type != room_type:
                continue
            if guests_count and room.max_guests < guests_count:
                continue
            is_free = True
            for res in self.reservations.values():
                if res.room_number == room.room_number:
                    if res.overlaps(check_in, check_out):
                        is_free = False
                        break

            if is_free:
                available.append(room)

        return available

    def book_room(self, guest_id, room_number, check_in, check_out, guests_count=1):

        if guest_id not in self.guests:
            raise ValueError("Guest ID not found.")
        if room_number not in self.rooms:
            raise ValueError("Room Number not found.")

        room = self.rooms[room_number]
        if room.max_guests < guests_count:
            raise ValueError(f"Room capacity exceeded. Max is {room.max_guests}.")


        for res in self.reservations.values():
            if res.room_number == room_number:
                if res.overlaps(check_in, check_out):
                    raise ValueError(f"Room {room_number} is not available for these dates.")


        res_id = str(uuid.uuid4())[:8]
        reservation = Reservation(res_id, guest_id, room_number, check_in, check_out)
        self.reservations[res_id] = reservation
        return reservation

    def cancel_reservation(self, reservation_id):
        if reservation_id not in self.reservations:
            raise ValueError("Reservation not found.")

        res = self.reservations[reservation_id]
        if res.status == "checked_out":
            raise ValueError("Cannot cancel a completed stay.")

        res.status = "cancelled"

    def check_in(self, reservation_id):
        if reservation_id not in self.reservations:
            raise ValueError("Reservation not found.")

        res = self.reservations[reservation_id]
        if res.status != "booked":
            raise ValueError(f"Cannot check-in. Current status is '{res.status}'.")

        res.status = "checked_in"

    def check_out(self, reservation_id):
        if reservation_id not in self.reservations:
            raise ValueError("Reservation not found.")

        res = self.reservations[reservation_id]
        if res.status != "checked_in":
            raise ValueError(f"Cannot check-out. Current status is '{res.status}'.")

        res.status = "checked_out"

    def list_reservations(self, filter_by_guest_id=None):
        results = []
        for res in self.reservations.values():
            if filter_by_guest_id and res.guest_id != filter_by_guest_id:
                continue
            results.append(res)
        return results

    def save(self, filename):
        data = {
            "rooms": [r.to_dict() for r in self.rooms.values()],
            "guests": [g.to_dict() for g in self.guests.values()],
            "reservations": [r.to_dict() for r in self.reservations.values()]
        }
        with open(filename, 'w') as f:
            json.dump(data, f, indent=4)
        print("Data saved successfully.")

    def load(self, filename):
        if not os.path.exists(filename):
            print("File not found.")
            return

        try:
            with open(filename, 'r') as f:
                data = json.load(f)


            self.rooms = {}
            self.guests = {}
            self.reservations = {}


            for r_data in data["rooms"]:
                room = Room.from_dict(r_data)
                self.rooms[room.room_number] = room


            for g_data in data["guests"]:
                guest = Guest.from_dict(g_data)
                self.guests[guest.guest_id] = guest


            for res_data in data["reservations"]:
                res = Reservation.from_dict(res_data)
                self.reservations[res.reservation_id] = res

            print("Data loaded successfully.")
        except json.JSONDecodeError:
            print("Error decoding JSON.")
        except KeyError as e:
            print(f"Data format error: missing key {e}")


# ==========================================
# 3. Console Helpers & Main Loop
# ==========================================

def get_date_input(prompt):
    while True:
        s = input(prompt)
        try:
            return datetime.date.fromisoformat(s)
        except ValueError:
            print("Invalid date format. Use YYYY-MM-DD.")

def get_int_input(prompt):
    while True:
        s = input(prompt)
        try:
            return int(s)
        except ValueError:
            print("Please enter a valid number.")

def get_float_input(prompt):
    while True:
        s = input(prompt)
        try:
            return float(s)
        except ValueError:
            print("Please enter a valid price.")

def main():
    hotel = Hotel()
    filename = "hotel_data.json"

    while True:
        print("\n=== HOTEL MANAGEMENT SYSTEM ===")
        print("1. Add Room")
        print("2. Register Guest")
        print("3. Search Available Rooms")
        print("4. Book Room")
        print("5. View Reservations")
        print("6. Cancel Reservation")
        print("7. Check-in")
        print("8. Check-out")
        print("9. Save to JSON")
        print("10. Load from JSON")
        print("11. Exit")

        choice = input("Select an option: ").strip()

        try:
            if choice == '1':
                r_num = get_int_input("Room Number: ")
                r_type = input("Type (single/double/suite): ").lower()
                price = get_float_input("Price per night: ")
                max_g = get_int_input("Max Guests: ")

                try:
                    r = Room(r_num, r_type, price, max_g)
                    hotel.add_room(r)
                    print(f"Room {r_num} added.")
                except ValueError as e:
                    print(f"Error: {e}")

            elif choice == '2':
                g_id = input("Guest ID (unique): ").strip()
                name = input("Name: ").strip()
                phone = input("Phone: ").strip()

                try:
                    g = Guest(g_id, name, phone)
                    hotel.register_guest(g)
                    print(f"Guest {name} registered.")
                except ValueError as e:
                    print(f"Error: {e}")

            elif choice == '3':
                cin = get_date_input("Check-in (YYYY-MM-DD): ")
                cout = get_date_input("Check-out (YYYY-MM-DD): ")

                print("Filters (press Enter to skip):")
                rtype = input("Room Type: ").strip().lower() or None
                cnt_str = input("Guest Count: ").strip()
                cnt = int(cnt_str) if cnt_str else None

                if cin >= cout:
                    print("Error: Check-out must be after check-in.")
                    continue

                available = hotel.search_available_rooms(cin, cout, rtype, cnt)
                if not available:
                    print("No rooms found.")
                else:
                    print("\n--- Available Rooms ---")
                    for r in available:
                        print(r)

            elif choice == '4':
                g_id = input("Guest ID: ").strip()
                r_num = get_int_input("Room Number: ")
                cin = get_date_input("Check-in (YYYY-MM-DD): ")
                cout = get_date_input("Check-out (YYYY-MM-DD): ")

                cnt_str = input("Number of Guests [1]: ").strip()
                cnt = int(cnt_str) if cnt_str else 1

                try:
                    res = hotel.book_room(g_id, r_num, cin, cout, cnt)
                    room = hotel.rooms[r_num]
                    cost = res.total_cost(room.price_per_night)
                    print(f"Booking Success! ID: {res.reservation_id}, Total Cost: ${cost:.2f}")
                except ValueError as e:
                    print(f"Booking Failed: {e}")

            elif choice == '5':
                filter_gid = input("Filter by Guest ID (Enter for all): ").strip() or None
                reservations = hotel.list_reservations(filter_gid)

                if not reservations:
                    print("No reservations found.")
                else:
                    print("\n--- Reservations ---")
                    for res in reservations:
                        print(res)

            elif choice == '6':
                rid = input("Reservation ID to cancel: ").strip()
                try:
                    hotel.cancel_reservation(rid)
                    print("Reservation cancelled.")
                except ValueError as e:
                    print(f"Error: {e}")

            elif choice == '7':
                rid = input("Reservation ID to Check-In: ").strip()
                try:
                    hotel.check_in(rid)
                    print("Check-in successful.")
                except ValueError as e:
                    print(f"Error: {e}")

            elif choice == '8':
                rid = input("Reservation ID to Check-Out: ").strip()
                try:
                    hotel.check_out(rid)
                    print("Check-out successful. Room is now free.")
                except ValueError as e:
                    print(f"Error: {e}")

            elif choice == '9':
                hotel.save(filename)

            elif choice == '10':
                hotel.load(filename)

            elif choice == '11':
                print("Goodbye!")
                break

            else:
                print("Invalid option. Please try again.")

        except Exception as e:
            print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```
